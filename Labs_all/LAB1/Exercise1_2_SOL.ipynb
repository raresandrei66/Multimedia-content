```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Create a pandas DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000]
}
```

create DataFrame

```python
df1 = pd.DataFrame(data)
```

display it

```python
df1
```


Accessing and modifying specific values in the DataFrame

Change Alice's age to 26

```python
df1.loc[0, 'Age'] = 26
```

Increase Bob's salary by 5000


```python
df1.loc[1, 'Salary'] += 5000

df1
```

Given data2 concatenate it with data1 and display the new results

```python
data2 = {
    'Name': ['Frank', 'Grace', 'Hannah', 'Ivy', 'Jack'],
    'Age': [30, 35, 40, 45, 50],
    'Salary': [75000, 85000, 95000, 105000, 115000]}

df2 = pd.DataFrame(data2)

combined_df = pd.concat([df2, df1], ignore_index=True)

combined_df
```

Sort the results by Salary, Age and Name

```python
combined_df.sort_values(by='Salary')
```

Plot Age vs. Salary

```python
combined_df.sort_values(by='Age')
```
